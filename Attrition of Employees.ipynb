<h5>Null values at NumCompaniesWorked has 19,TotalWorkingYears has 9,EnvironmentSatisfaction has 25,JobSatisfaction has 20,WorkLifeBalance has 38 </h5>

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

<h5> Droped Null Values Now Data Is Cleaned </h5>

```python
l=data['EmployeeCount'].count()
print('employee count:',l)
```

```python
data1=data[data['Attrition']=='Yes']
k=data['Attrition']=='Yes'
k.astype(int)
c=0
for i in k:
    if i==1:
        c+=1
print('Attrition Count:',c)
```

```python
data1
```

```python
Active=l-c
print('Active Employee:',Active)
```

```python
print('Attrition rate:',round((c/l)*100,2))
```

```python
gender=data1.groupby('Gender').count()
sns.barplot(gender.EmployeeID);
```

```python
pie1=data1.groupby('Department').count()
plt.pie(pie1.EmployeeID,labels=pie1.index,autopct='%1.2f%%');
```

```python
plt.figure(figsize=(20,10))
x=data1.groupby(['Age']).count()
sns.barplot(x.EmployeeID);
```

```python
plt.figure(figsize=(20,10))
education=data1.groupby('EducationField').count()
sns.barplot(education.EmployeeID);
```
